This notebook explores a single 4 band (blue, green, red, NIR) PlanetScope scene

```python
import rasterio
from rasterio import plot
import matplotlib.pyplot as plt
```

```python
# Use Rasterio to open the image
satdat1 = rasterio.open("sample.tif")
```

```python
#show the number of bands
satdat1.count
```

```python
#show the width of the raster
satdat1.width
```

```python
#show the height of the raster
satdat1.height
```

```python
#show the raster CRS
satdat1.crs
```

WGS 84 / UTM zone 11N 

```python
#plot the image using the 'twilight' color map
plot.show(satdat1, cmap='twilight')
```

```python
#plot each band of the image individually
fig, (axr, axg, axb, axn) = plt.subplots(1,4, figsize=(20,5))
plot.show((satdat1, 3), ax=axr, cmap='Reds', title='red channel')
plot.show((satdat1, 2), ax=axg, cmap='Greens', title='green channel')
plot.show((satdat1, 1), ax=axb, cmap='Blues', title='blue channel')
plot.show((satdat1, 4), ax=axn, cmap='RdPu', title='NIR channel')
```
